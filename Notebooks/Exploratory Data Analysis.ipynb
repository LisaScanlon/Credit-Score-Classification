```python
import pandas as pd
data = pd.read_csv('cleaned_train.csv')
```

```python
import matplotlib.pyplot as plt

# Count occurrences of each category in the "Credit_Score" column
credit_score_counts = data['Credit_Score'].value_counts()

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(credit_score_counts, labels=credit_score_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Credit Scores')
plt.axis('equal')  # Equal argument makes the pie chart get drawn in a circle
plt.show()
```

```python
# Extracting data for visualization
age = data['Age']
num_of_bank_accounts = data['Num_Bank_Accounts']

# Creating scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(age, num_of_bank_accounts, color='blue', alpha=0.5)
plt.title('Age vs Number of Bank Accounts')
plt.xlabel('Age')
plt.ylabel('Number of Bank Accounts')
plt.grid(True)
plt.show()
```

```python
# Count occurrences of each category in the "Credit_Score" column
credit_score_counts = data['Age'].value_counts()

# Plot pie chart
plt.figure(figsize=(15, 15))
plt.pie(credit_score_counts, labels=credit_score_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Age')
plt.axis('equal')  # Equal argument makes the pie chart get drawn in a circle
plt.show()
```

```python
# Bar Plot of Number of Bank Accounts
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.countplot(data['Num_Bank_Accounts'])
plt.title('Number of Bank Accounts')
plt.xlabel('Number of Bank Accounts')
plt.ylabel('Count')
plt.show()
```

```python
# Count occurrences of each category in the "Credit_Score" column
credit_score_counts = data['Occupation'].value_counts()

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(credit_score_counts, labels=credit_score_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Occupation')
plt.axis('equal')  # Equal argument makes the pie chart get drawn in a circle
plt.show()
```

```python
# Bar Plot of Occupation vs Annual Income
occupation_income = data.groupby('Occupation')['Annual_Income'].mean().sort_values(ascending=False).reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Occupation', y='Annual_Income', data=data, palette='pastel')
plt.title('Average Annual Income by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Average Annual Income')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create scatter plot or credit utilization rate and interest rate
plt.figure(figsize=(15, 15))
plt.scatter(data['Credit_Utilization_Ratio'], data['Interest_Rate'], color='orange', alpha=0.5)
plt.title('Scatter Plot of Credit Utilization Ratio vs. Interest Rate')
plt.xlabel('Credit Utilization Ratio')
plt.ylabel('Interest Rate')
plt.grid(True)
plt.show()
```
